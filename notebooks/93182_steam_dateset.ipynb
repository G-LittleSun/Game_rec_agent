# 特征分类与数据清洗

## 步骤1: 特征分类定义

```python
# 🔴 核心特征(必须) - 推荐系统的基础
core_features = [
    'AppID',           # 唯一标识
    'Name',            # 游戏名称
    'About the game',  # 游戏描述(语义搜索核心)
    'Genres',          # 类型
    'Tags',            # 标签（缺失太多）
    'Categories',      # 分类
]

# 🟡 重要特征(强烈推荐) - 用于排序和筛选
important_features = [
    'Price',           # 价格筛选
    'Positive',        # 好评数
    'Negative',        # 差评数
    'Recommendations', # 推荐数
    'Release date',    # 发布时间
]

# 🟢 辅助特征(可选) - 提供额外信息
optional_features = [
    'Developers',      # 开发商
    'Publishers',      # 发行商
    'Estimated owners',# 热度指标
    'Average playtime forever',  # 所有时间平均时长，可玩性指标
    'Metacritic score',# 专业评分
    'Windows', 'Mac', 'Linux',  # 平台
    'Required age', #年龄限制
    'Peak CCU', #历史玩家在线峰值
    'Supported languages', # 支持语言
    'Average playtime two weeks', # 近两周的平均游戏时长
    'Median playtime forever', # 所有时间平均游戏时长中位数
    'Median playtime two weeks', # 近两周的平均游戏时长中位数
]

# 🔵 可忽略特征 - 对推荐系统帮助不大
ignore_features = [
    'Support url', 'Support email', 'Website',
    'Header image', 'Screenshots', 'Movies',
    'Metacritic url', 'Notes', 'Reviews',
    'User score', 'Score rank', 'DLC count',  
    'Full audio languages', 
    'Achievements'
]

# 合并需要保留的特征
selected_features = core_features + important_features + optional_features

print("📊 特征分类统计:")
print(f"核心特征: {len(core_features)}个")
print(f"重要特征: {len(important_features)}个")
print(f"辅助特征: {len(optional_features)}个")
print(f"可忽略特征: {len(ignore_features)}个")
print(f"\n总共保留: {len(selected_features)}个特征")
print(f"将忽略: {len(ignore_features)}个特征")
print(f"\n保留的特征列表:")
for i, feat in enumerate(selected_features, 1):
    print(f"{i}. {feat}")
```

```python
# 创建筛选后的数据框副本
df_clean = df[selected_features].copy()

print(f"✅ 特征筛选完成!")
print(f"原始数据形状: {df.shape}")
print(f"筛选后形状: {df_clean.shape}")
print(f"减少了 {df.shape[1] - df_clean.shape[1]} 列")
```

## 步骤2: 处理重复数据

```python
# 检查重复数据
print("🔍 检查重复数据:")
print(f"完全重复的行数: {df_clean.duplicated().sum()}")
print(f"基于AppID的重复数: {df_clean['AppID'].duplicated().sum()}")
```

## 步骤3: 分析缺失值情况

```python
# 计算缺失值统计
print("🔍 缺失值详细分析:")
print("="*80)

missing_analysis = pd.DataFrame({
    '缺失数量': df_clean.isnull().sum(),
    '缺失比例(%)': (df_clean.isnull().sum() / len(df_clean) * 100).round(2),
    '非缺失数量': df_clean.notnull().sum()
})

# 只显示有缺失值的列
missing_analysis = missing_analysis[missing_analysis['缺失数量'] > 0].sort_values('缺失比例(%)', ascending=False)

if len(missing_analysis) > 0:
    print(missing_analysis)
    print("\n" + "="*80)
    
    # 按特征类型分类统计缺失情况
    print("\n按特征类型分类的缺失情况:")
    for feat in missing_analysis.index:
        if feat in core_features:
            print(f"🔴 核心特征 - {feat}: 缺失{missing_analysis.loc[feat, '缺失比例(%)']}%")
        elif feat in important_features:
            print(f"🟡 重要特征 - {feat}: 缺失{missing_analysis.loc[feat, '缺失比例(%)']}%")
        elif feat in optional_features:
            print(f"🟢 辅助特征 - {feat}: 缺失{missing_analysis.loc[feat, '缺失比例(%)']}%")
else:
    print("✅ 没有发现缺失值!")
```

## 步骤4: 处理缺失值

```python
# 缺失值处理策略
print("🛠️ 缺失值处理策略:")
print("="*80)

# 策略1: 核心特征 - 删除缺失的行(这些字段对推荐系统至关重要)
print("\n1️⃣ 核心特征处理(删除缺失行):")
core_missing_before = df_clean[core_features].isnull().sum().sum()

# Tags缺失太多,单独处理为'Unknown'而不是删除
if 'Tags' in df_clean.columns:
    tags_missing = df_clean['Tags'].isnull().sum()
    if tags_missing > 0:
        print(f"   - Tags: 缺失 {tags_missing} 行 -> 填充为 'Unknown' (缺失太多,不删除)")
        df_clean['Tags'] = df_clean['Tags'].fillna('Unknown')

# 其他核心特征缺失则删除行
for feat in core_features:
    if feat == 'Tags':  # Tags已经处理过了
        continue
    missing_count = df_clean[feat].isnull().sum()
    if missing_count > 0:
        print(f"   - {feat}: 缺失 {missing_count} 行 -> 将删除")
        df_clean = df_clean[df_clean[feat].notna()]
        
print(f"   核心特征缺失总数: {core_missing_before}")
print(f"   当前数据行数: {len(df_clean)}")

# 策略2: 重要特征 - 合理填充
print("\n2️⃣ 重要特征处理(填充缺失值):")

# Price - 填充为0(免费游戏)
if 'Price' in df_clean.columns:
    price_missing = df_clean['Price'].isnull().sum()
    if price_missing > 0:
        df_clean['Price'] = df_clean['Price'].fillna(0)
        print(f"   - Price: 缺失 {price_missing} 行 -> 填充为 0 (免费游戏)")

# Positive/Negative - 填充为0(无评价)
for feat in ['Positive', 'Negative']:
    if feat in df_clean.columns:
        missing = df_clean[feat].isnull().sum()
        if missing > 0:
            df_clean[feat] = df_clean[feat].fillna(0)
            print(f"   - {feat}: 缺失 {missing} 行 -> 填充为 0 (无评价)")

# Recommendations - 填充为0
if 'Recommendations' in df_clean.columns:
    rec_missing = df_clean['Recommendations'].isnull().sum()
    if rec_missing > 0:
        df_clean['Recommendations'] = df_clean['Recommendations'].fillna(0)
        print(f"   - Recommendations: 缺失 {rec_missing} 行 -> 填充为 0")

# Release date - 保留缺失(后续处理)
if 'Release date' in df_clean.columns:
    date_missing = df_clean['Release date'].isnull().sum()
    if date_missing > 0:
        print(f"   - Release date: 缺失 {date_missing} 行 -> 保留(后续日期转换时处理)")

# 策略3: 辅助特征 - 填充默认值或保留
print("\n3️⃣ 辅助特征处理(填充默认值):")

# Developers/Publishers - 填充为'Unknown'
for feat in ['Developers', 'Publishers']:
    if feat in df_clean.columns:
        missing = df_clean[feat].isnull().sum()
        if missing > 0:
            df_clean[feat] = df_clean[feat].fillna('Unknown')
            print(f"   - {feat}: 缺失 {missing} 行 -> 填充为 'Unknown'")

# Estimated owners - 填充为'0 - 0'
if 'Estimated owners' in df_clean.columns:
    owners_missing = df_clean['Estimated owners'].isnull().sum()
    if owners_missing > 0:
        df_clean['Estimated owners'] = df_clean['Estimated owners'].fillna('0 - 0')
        print(f"   - Estimated owners: 缺失 {owners_missing} 行 -> 填充为 '0 - 0'")

# 游戏时长相关字段 - 填充为0
for playtime_feat in ['Average playtime forever', 'Average playtime two weeks', 
                       'Median playtime forever', 'Median playtime two weeks']:
    if playtime_feat in df_clean.columns:
        playtime_missing = df_clean[playtime_feat].isnull().sum()
        if playtime_missing > 0:
            df_clean[playtime_feat] = df_clean[playtime_feat].fillna(0)
            print(f"   - {playtime_feat}: 缺失 {playtime_missing} 行 -> 填充为 0")

# Peak CCU - 填充为0
if 'Peak CCU' in df_clean.columns:
    peak_missing = df_clean['Peak CCU'].isnull().sum()
    if peak_missing > 0:
        df_clean['Peak CCU'] = df_clean['Peak CCU'].fillna(0)
        print(f"   - Peak CCU: 缺失 {peak_missing} 行 -> 填充为 0")

# Required age - 填充为0
if 'Required age' in df_clean.columns:
    age_missing = df_clean['Required age'].isnull().sum()
    if age_missing > 0:
        df_clean['Required age'] = df_clean['Required age'].fillna(0)
        print(f"   - Required age: 缺失 {age_missing} 行 -> 填充为 0")

# Supported languages - 填充为'Unknown'
if 'Supported languages' in df_clean.columns:
    lang_missing = df_clean['Supported languages'].isnull().sum()
    if lang_missing > 0:
        df_clean['Supported languages'] = df_clean['Supported languages'].fillna('Unknown')
        print(f"   - Supported languages: 缺失 {lang_missing} 行 -> 填充为 'Unknown'")

# Metacritic score - 填充为0(无评分)
if 'Metacritic score' in df_clean.columns:
    meta_missing = df_clean['Metacritic score'].isnull().sum()
    if meta_missing > 0:
        df_clean['Metacritic score'] = df_clean['Metacritic score'].fillna(0)
        print(f"   - Metacritic score: 缺失 {meta_missing} 行 -> 填充为 0 (无评分)")

# Windows/Mac/Linux - 填充为False
for platform in ['Windows', 'Mac', 'Linux']:
    if platform in df_clean.columns:
        platform_missing = df_clean[platform].isnull().sum()
        if platform_missing > 0:
            df_clean[platform] = df_clean[platform].fillna(False)
            print(f"   - {platform}: 缺失 {platform_missing} 行 -> 填充为 False")

print("\n" + "="*80)
print(f"✅ 缺失值处理完成! 当前数据行数: {len(df_clean)}")
```

## 步骤5: 数据类型转换和规范化

```python
print("🔄 数据类型转换和规范化:")
print("="*80)

# 1. 转换日期格式 (Release date)
print("\n1️⃣ 转换发布日期格式:")
if 'Release date' in df_clean.columns:
    print(f"   原始数据类型: {df_clean['Release date'].dtype}")
    print(f"   样例数据: {df_clean['Release date'].head(3).tolist()}")
    
    # 转换为datetime格式,错误的值设为NaT
    df_clean['Release date'] = pd.to_datetime(df_clean['Release date'], errors='coerce')
    
    # 统计转换失败的数量
    failed_conversion = df_clean['Release date'].isnull().sum()
    print(f"   转换后数据类型: {df_clean['Release date'].dtype}")
    print(f"   转换失败/缺失: {failed_conversion} 行")
    
    # 为缺失的日期填充一个默认值(如1970-01-01)或保持NaN
    # 这里我们保持NaT,后续使用时再决定如何处理
    print(f"   ✅ 日期转换完成")

# 2. 规范化价格字段 (Price)
print("\n2️⃣ 规范化价格字段:")
if 'Price' in df_clean.columns:
    print(f"   原始数据类型: {df_clean['Price'].dtype}")
    
    # 确保价格是数值型
    df_clean['Price'] = pd.to_numeric(df_clean['Price'], errors='coerce')
    
    # 处理异常价格(负值设为0)
    negative_prices = (df_clean['Price'] < 0).sum()
    if negative_prices > 0:
        print(f"   发现 {negative_prices} 个负价格,已设为0")
        df_clean.loc[df_clean['Price'] < 0, 'Price'] = 0
    
    # 填充转换失败的值为0
    df_clean['Price'] = df_clean['Price'].fillna(0)
    
    print(f"   价格范围: ${df_clean['Price'].min():.2f} - ${df_clean['Price'].max():.2f}")
    print(f"   平均价格: ${df_clean['Price'].mean():.2f}")
    print(f"   ✅ 价格规范化完成")

# 3. 转换评分字段为整数类1型
print("\n3️⃣ 转换评分字段:")
for col in ['Positive', 'Negative', 'Recommendations']:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce').fillna(0).astype(int)
        print(f"   - {col}: 转换为整数类型 -> 范围 [{df_clean[col].min()}, {df_clean[col].max()}]")

# 4. 处理混合类型的列 - Estimated owners
print("\n4️⃣ 处理Estimated owners字段:")
if 'Estimated owners' in df_clean.columns:
    print(f"   原始数据类型: {df_clean['Estimated owners'].dtype}")
    print(f"   样例数据: {df_clean['Estimated owners'].head(3).tolist()}")
    
    # 将字符串转换为字符串类型(统一格式)
    df_clean['Estimated owners'] = df_clean['Estimated owners'].astype(str)
    print(f"   ✅ 转换为字符串类型")

# 5. 转换平台字段为布尔型
print("\n5️⃣ 转换平台支持字段:")
for platform in ['Windows', 'Mac', 'Linux']:
    if platform in df_clean.columns:
        # 转换为布尔类型
        df_clean[platform] = df_clean[platform].astype(bool)
        support_count = df_clean[platform].sum()
        print(f"   - {platform}: {support_count} 个游戏支持 ({support_count/len(df_clean)*100:.1f}%)")

# 6. 转换Metacritic score为数值型
print("\n6️⃣ 转换Metacritic score:")
if 'Metacritic score' in df_clean.columns:
    df_clean['Metacritic score'] = pd.to_numeric(df_clean['Metacritic score'], errors='coerce').fillna(0).astype(int)
    non_zero = (df_clean['Metacritic score'] > 0).sum()
    print(f"   有Metacritic评分的游戏: {non_zero} 个 ({non_zero/len(df_clean)*100:.1f}%)")
    if non_zero > 0:
        print(f"   评分范围: {df_clean[df_clean['Metacritic score']>0]['Metacritic score'].min()} - {df_clean['Metacritic score'].max()}")
    print(f"   ✅ 转换完成")

# 7. 转换游戏时长字段为数值型
print("\n7️⃣ 转换游戏时长字段:")
for playtime_col in ['Average playtime forever', 'Average playtime two weeks',
                      'Median playtime forever', 'Median playtime two weeks']:
    if playtime_col in df_clean.columns:
        df_clean[playtime_col] = pd.to_numeric(df_clean[playtime_col], errors='coerce').fillna(0)
        print(f"   - {playtime_col}: 平均 {df_clean[playtime_col].mean():.2f} 分钟")

# 8. 转换其他辅助字段
print("\n8️⃣ 转换其他辅助字段:")
# Peak CCU
if 'Peak CCU' in df_clean.columns:
    df_clean['Peak CCU'] = pd.to_numeric(df_clean['Peak CCU'], errors='coerce').fillna(0).astype(int)
    print(f"   - Peak CCU: 最大值 {df_clean['Peak CCU'].max():,}")

# Required age
if 'Required age' in df_clean.columns:
    df_clean['Required age'] = pd.to_numeric(df_clean['Required age'], errors='coerce').fillna(0).astype(int)
    print(f"   - Required age: 范围 {df_clean['Required age'].min()} - {df_clean['Required age'].max()}")

# Supported languages - 保持字符串类型
if 'Supported languages' in df_clean.columns:
    df_clean['Supported languages'] = df_clean['Supported languages'].astype(str)
    print(f"   - Supported languages: 转换为字符串类型")

print("\n" + "="*80)
print(f"✅ 数据类型转换完成!")
```

## 步骤6: 验证清洗结果

```python
print("📋 数据清洗结果验证:")
print("="*80)

# 1. 数据形状
print(f"\n1️⃣ 数据形状:")
print(f"   行数: {df_clean.shape[0]:,}")
print(f"   列数: {df_clean.shape[1]}")

# 2. 数据类型
print(f"\n2️⃣ 数据类型分布:")
print(df_clean.dtypes.value_counts())

# 3. 缺失值最终统计
print(f"\n3️⃣ 剩余缺失值:")
remaining_missing = df_clean.isnull().sum().sum()
if remaining_missing > 0:
    print(f"   总缺失值: {remaining_missing}")
    missing_cols = df_clean.isnull().sum()
    missing_cols = missing_cols[missing_cols > 0]
    for col, count in missing_cols.items():
        print(f"   - {col}: {count} ({count/len(df_clean)*100:.2f}%)")
else:
    print(f"   ✅ 无缺失值!")

# 4. 重复检查
print(f"\n4️⃣ 重复数据:")
print(f"   完全重复: {df_clean.duplicated().sum()}")
print(f"   AppID重复: {df_clean['AppID'].duplicated().sum()}")

# 5. 查看清洗后的样例数据
print(f"\n5️⃣ 清洗后样例数据:")
print(df_clean.head(3))

# 6. 核心字段数据质量
print(f"\n6️⃣ 核心字段数据质量检查:")
for feat in core_features:
    if feat in df_clean.columns:
        non_null = df_clean[feat].notna().sum()
        print(f"   - {feat}: {non_null}/{len(df_clean)} ({non_null/len(df_clean)*100:.2f}% 完整)")

print("\n" + "="*80)
print(f"✅ 数据清洗验证完成!")
```

## 步骤7: 保存清洗后的数据

```python
# 保存清洗后的数据到processed文件夹
import os

# 确保processed文件夹存在
output_dir = "../data/processed"
os.makedirs(output_dir, exist_ok=True)

# 保存为CSV文件
output_path = os.path.join(output_dir, "steam_games_cleaned.csv")
df_clean.to_csv(output_path, index=False, encoding='utf-8')

print(f"💾 数据保存成功!")
print(f"   保存路径: {output_path}")
print(f"   保存行数: {len(df_clean):,}")
print(f"   保存列数: {df_clean.shape[1]}")
print(f"   文件大小: {os.path.getsize(output_path) / 1024 / 1024:.2f} MB")

# 也保存一个parquet格式(更高效)
output_path_parquet = os.path.join(output_dir, "steam_games_cleaned.parquet")
df_clean.to_parquet(output_path_parquet, index=False)
print(f"\n   同时保存为Parquet格式: {output_path_parquet}")
print(f"   Parquet文件大小: {os.path.getsize(output_path_parquet) / 1024 / 1024:.2f} MB")

print("\n✅ 数据清洗流程全部完成!")
```
