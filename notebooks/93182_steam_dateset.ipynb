## æ­¥éª¤1: ç‰¹å¾åˆ†ç±»å®šä¹‰

```python
# ğŸ”´ æ ¸å¿ƒç‰¹å¾(å¿…é¡») - æ¨èç³»ç»Ÿçš„åŸºç¡€
core_features = [
    'AppID',           # å”¯ä¸€æ ‡è¯†
    'Name',            # æ¸¸æˆåç§°
    'About the game',  # æ¸¸æˆæè¿°(è¯­ä¹‰æœç´¢æ ¸å¿ƒ)
    'Genres',          # ç±»å‹
    'Tags',            # æ ‡ç­¾ï¼ˆç¼ºå¤±å¤ªå¤šï¼‰
    'Categories',      # åˆ†ç±»
]

# ğŸŸ¡ é‡è¦ç‰¹å¾(å¼ºçƒˆæ¨è) - ç”¨äºæ’åºå’Œç­›é€‰
important_features = [
    'Price',           # ä»·æ ¼ç­›é€‰
    'Positive',        # å¥½è¯„æ•°
    'Negative',        # å·®è¯„æ•°
    'Recommendations', # æ¨èæ•°
    'Release date',    # å‘å¸ƒæ—¶é—´
]

# ğŸŸ¢ è¾…åŠ©ç‰¹å¾(å¯é€‰) - æä¾›é¢å¤–ä¿¡æ¯
optional_features = [
    'Developers',      # å¼€å‘å•†
    'Publishers',      # å‘è¡Œå•†
    'Estimated owners',# çƒ­åº¦æŒ‡æ ‡
    'Average playtime forever',  # æ‰€æœ‰æ—¶é—´å¹³å‡æ—¶é•¿ï¼Œå¯ç©æ€§æŒ‡æ ‡
    'Metacritic score',# ä¸“ä¸šè¯„åˆ†
    'Windows', 'Mac', 'Linux',  # å¹³å°
    'Required age', #å¹´é¾„é™åˆ¶
    'Peak CCU', #å†å²ç©å®¶åœ¨çº¿å³°å€¼
    'Supported languages', # æ”¯æŒè¯­è¨€
    'Average playtime two weeks', # è¿‘ä¸¤å‘¨çš„å¹³å‡æ¸¸æˆæ—¶é•¿
    'Median playtime forever', # æ‰€æœ‰æ—¶é—´å¹³å‡æ¸¸æˆæ—¶é•¿ä¸­ä½æ•°
    'Median playtime two weeks', # è¿‘ä¸¤å‘¨çš„å¹³å‡æ¸¸æˆæ—¶é•¿ä¸­ä½æ•°
]

# ğŸ”µ å¯å¿½ç•¥ç‰¹å¾ - å¯¹æ¨èç³»ç»Ÿå¸®åŠ©ä¸å¤§
ignore_features = [
    'Support url', 'Support email', 'Website',
    'Header image', 'Screenshots', 'Movies',
    'Metacritic url', 'Notes', 'Reviews',
    'User score', 'Score rank', 'DLC count',  
    'Full audio languages', 
    'Achievements'
]

# åˆå¹¶éœ€è¦ä¿ç•™çš„ç‰¹å¾
selected_features = core_features + important_features + optional_features

print("ğŸ“Š ç‰¹å¾åˆ†ç±»ç»Ÿè®¡:")
print(f"æ ¸å¿ƒç‰¹å¾: {len(core_features)}ä¸ª")
print(f"é‡è¦ç‰¹å¾: {len(important_features)}ä¸ª")
print(f"è¾…åŠ©ç‰¹å¾: {len(optional_features)}ä¸ª")
print(f"å¯å¿½ç•¥ç‰¹å¾: {len(ignore_features)}ä¸ª")
print(f"\næ€»å…±ä¿ç•™: {len(selected_features)}ä¸ªç‰¹å¾")
print(f"å°†å¿½ç•¥: {len(ignore_features)}ä¸ªç‰¹å¾")
print(f"\nä¿ç•™çš„ç‰¹å¾åˆ—è¡¨:")
for i, feat in enumerate(selected_features, 1):
    print(f"{i}. {feat}")
```

```python
# åˆ›å»ºç­›é€‰åçš„æ•°æ®æ¡†å‰¯æœ¬
df_clean = df[selected_features].copy()

print(f"âœ… ç‰¹å¾ç­›é€‰å®Œæˆ!")
print(f"åŸå§‹æ•°æ®å½¢çŠ¶: {df.shape}")
print(f"ç­›é€‰åå½¢çŠ¶: {df_clean.shape}")
print(f"å‡å°‘äº† {df.shape[1] - df_clean.shape[1]} åˆ—")
```

## æ­¥éª¤3: åˆ†æç¼ºå¤±å€¼æƒ…å†µ

```python
# è®¡ç®—ç¼ºå¤±å€¼ç»Ÿè®¡
print("ğŸ” ç¼ºå¤±å€¼è¯¦ç»†åˆ†æ:")
print("="*80)

missing_analysis = pd.DataFrame({
    'ç¼ºå¤±æ•°é‡': df_clean.isnull().sum(),
    'ç¼ºå¤±æ¯”ä¾‹(%)': (df_clean.isnull().sum() / len(df_clean) * 100).round(2),
    'éç¼ºå¤±æ•°é‡': df_clean.notnull().sum()
})

# åªæ˜¾ç¤ºæœ‰ç¼ºå¤±å€¼çš„åˆ—
missing_analysis = missing_analysis[missing_analysis['ç¼ºå¤±æ•°é‡'] > 0].sort_values('ç¼ºå¤±æ¯”ä¾‹(%)', ascending=False)

if len(missing_analysis) > 0:
    print(missing_analysis)
    print("\n" + "="*80)
    
    # æŒ‰ç‰¹å¾ç±»å‹åˆ†ç±»ç»Ÿè®¡ç¼ºå¤±æƒ…å†µ
    print("\næŒ‰ç‰¹å¾ç±»å‹åˆ†ç±»çš„ç¼ºå¤±æƒ…å†µ:")
    for feat in missing_analysis.index:
        if feat in core_features:
            print(f"ğŸ”´ æ ¸å¿ƒç‰¹å¾ - {feat}: ç¼ºå¤±{missing_analysis.loc[feat, 'ç¼ºå¤±æ¯”ä¾‹(%)']}%")
        elif feat in important_features:
            print(f"ğŸŸ¡ é‡è¦ç‰¹å¾ - {feat}: ç¼ºå¤±{missing_analysis.loc[feat, 'ç¼ºå¤±æ¯”ä¾‹(%)']}%")
        elif feat in optional_features:
            print(f"ğŸŸ¢ è¾…åŠ©ç‰¹å¾ - {feat}: ç¼ºå¤±{missing_analysis.loc[feat, 'ç¼ºå¤±æ¯”ä¾‹(%)']}%")
else:
    print("âœ… æ²¡æœ‰å‘ç°ç¼ºå¤±å€¼!")
```

## æ­¥éª¤4: å¤„ç†ç¼ºå¤±å€¼

```python
# ç¼ºå¤±å€¼å¤„ç†ç­–ç•¥
print("ğŸ› ï¸ ç¼ºå¤±å€¼å¤„ç†ç­–ç•¥:")
print("="*80)

# ç­–ç•¥1: æ ¸å¿ƒç‰¹å¾ - åˆ é™¤ç¼ºå¤±çš„è¡Œ(è¿™äº›å­—æ®µå¯¹æ¨èç³»ç»Ÿè‡³å…³é‡è¦)
print("\n1ï¸âƒ£ æ ¸å¿ƒç‰¹å¾å¤„ç†(åˆ é™¤ç¼ºå¤±è¡Œ):")
core_missing_before = df_clean[core_features].isnull().sum().sum()

# Tagsç¼ºå¤±å¤ªå¤š,å•ç‹¬å¤„ç†ä¸º'Unknown'è€Œä¸æ˜¯åˆ é™¤
if 'Tags' in df_clean.columns:
    tags_missing = df_clean['Tags'].isnull().sum()
    if tags_missing > 0:
        print(f"   - Tags: ç¼ºå¤± {tags_missing} è¡Œ -> å¡«å……ä¸º 'Unknown' (ç¼ºå¤±å¤ªå¤š,ä¸åˆ é™¤)")
        df_clean['Tags'] = df_clean['Tags'].fillna('Unknown')

# å…¶ä»–æ ¸å¿ƒç‰¹å¾ç¼ºå¤±åˆ™åˆ é™¤è¡Œ
for feat in core_features:
    if feat == 'Tags':  # Tagså·²ç»å¤„ç†è¿‡äº†
        continue
    missing_count = df_clean[feat].isnull().sum()
    if missing_count > 0:
        print(f"   - {feat}: ç¼ºå¤± {missing_count} è¡Œ -> å°†åˆ é™¤")
        df_clean = df_clean[df_clean[feat].notna()]
        
print(f"   æ ¸å¿ƒç‰¹å¾ç¼ºå¤±æ€»æ•°: {core_missing_before}")
print(f"   å½“å‰æ•°æ®è¡Œæ•°: {len(df_clean)}")

# ç­–ç•¥2: é‡è¦ç‰¹å¾ - åˆç†å¡«å……
print("\n2ï¸âƒ£ é‡è¦ç‰¹å¾å¤„ç†(å¡«å……ç¼ºå¤±å€¼):")

# Price - å¡«å……ä¸º0(å…è´¹æ¸¸æˆ)
if 'Price' in df_clean.columns:
    price_missing = df_clean['Price'].isnull().sum()
    if price_missing > 0:
        df_clean['Price'] = df_clean['Price'].fillna(0)
        print(f"   - Price: ç¼ºå¤± {price_missing} è¡Œ -> å¡«å……ä¸º 0 (å…è´¹æ¸¸æˆ)")

# Positive/Negative - å¡«å……ä¸º0(æ— è¯„ä»·)
for feat in ['Positive', 'Negative']:
    if feat in df_clean.columns:
        missing = df_clean[feat].isnull().sum()
        if missing > 0:
            df_clean[feat] = df_clean[feat].fillna(0)
            print(f"   - {feat}: ç¼ºå¤± {missing} è¡Œ -> å¡«å……ä¸º 0 (æ— è¯„ä»·)")

# Recommendations - å¡«å……ä¸º0
if 'Recommendations' in df_clean.columns:
    rec_missing = df_clean['Recommendations'].isnull().sum()
    if rec_missing > 0:
        df_clean['Recommendations'] = df_clean['Recommendations'].fillna(0)
        print(f"   - Recommendations: ç¼ºå¤± {rec_missing} è¡Œ -> å¡«å……ä¸º 0")

# Release date - ä¿ç•™ç¼ºå¤±(åç»­å¤„ç†)
if 'Release date' in df_clean.columns:
    date_missing = df_clean['Release date'].isnull().sum()
    if date_missing > 0:
        print(f"   - Release date: ç¼ºå¤± {date_missing} è¡Œ -> ä¿ç•™(åç»­æ—¥æœŸè½¬æ¢æ—¶å¤„ç†)")

# ç­–ç•¥3: è¾…åŠ©ç‰¹å¾ - å¡«å……é»˜è®¤å€¼æˆ–ä¿ç•™
print("\n3ï¸âƒ£ è¾…åŠ©ç‰¹å¾å¤„ç†(å¡«å……é»˜è®¤å€¼):")

# Developers/Publishers - å¡«å……ä¸º'Unknown'
for feat in ['Developers', 'Publishers']:
    if feat in df_clean.columns:
        missing = df_clean[feat].isnull().sum()
        if missing > 0:
            df_clean[feat] = df_clean[feat].fillna('Unknown')
            print(f"   - {feat}: ç¼ºå¤± {missing} è¡Œ -> å¡«å……ä¸º 'Unknown'")

# Estimated owners - å¡«å……ä¸º'0 - 0'
if 'Estimated owners' in df_clean.columns:
    owners_missing = df_clean['Estimated owners'].isnull().sum()
    if owners_missing > 0:
        df_clean['Estimated owners'] = df_clean['Estimated owners'].fillna('0 - 0')
        print(f"   - Estimated owners: ç¼ºå¤± {owners_missing} è¡Œ -> å¡«å……ä¸º '0 - 0'")

# æ¸¸æˆæ—¶é•¿ç›¸å…³å­—æ®µ - å¡«å……ä¸º0
for playtime_feat in ['Average playtime forever', 'Average playtime two weeks', 
                       'Median playtime forever', 'Median playtime two weeks']:
    if playtime_feat in df_clean.columns:
        playtime_missing = df_clean[playtime_feat].isnull().sum()
        if playtime_missing > 0:
            df_clean[playtime_feat] = df_clean[playtime_feat].fillna(0)
            print(f"   - {playtime_feat}: ç¼ºå¤± {playtime_missing} è¡Œ -> å¡«å……ä¸º 0")

# Peak CCU - å¡«å……ä¸º0
if 'Peak CCU' in df_clean.columns:
    peak_missing = df_clean['Peak CCU'].isnull().sum()
    if peak_missing > 0:
        df_clean['Peak CCU'] = df_clean['Peak CCU'].fillna(0)
        print(f"   - Peak CCU: ç¼ºå¤± {peak_missing} è¡Œ -> å¡«å……ä¸º 0")

# Required age - å¡«å……ä¸º0
if 'Required age' in df_clean.columns:
    age_missing = df_clean['Required age'].isnull().sum()
    if age_missing > 0:
        df_clean['Required age'] = df_clean['Required age'].fillna(0)
        print(f"   - Required age: ç¼ºå¤± {age_missing} è¡Œ -> å¡«å……ä¸º 0")

# Supported languages - å¡«å……ä¸º'Unknown'
if 'Supported languages' in df_clean.columns:
    lang_missing = df_clean['Supported languages'].isnull().sum()
    if lang_missing > 0:
        df_clean['Supported languages'] = df_clean['Supported languages'].fillna('Unknown')
        print(f"   - Supported languages: ç¼ºå¤± {lang_missing} è¡Œ -> å¡«å……ä¸º 'Unknown'")

# Metacritic score - å¡«å……ä¸º0(æ— è¯„åˆ†)
if 'Metacritic score' in df_clean.columns:
    meta_missing = df_clean['Metacritic score'].isnull().sum()
    if meta_missing > 0:
        df_clean['Metacritic score'] = df_clean['Metacritic score'].fillna(0)
        print(f"   - Metacritic score: ç¼ºå¤± {meta_missing} è¡Œ -> å¡«å……ä¸º 0 (æ— è¯„åˆ†)")

# Windows/Mac/Linux - å¡«å……ä¸ºFalse
for platform in ['Windows', 'Mac', 'Linux']:
    if platform in df_clean.columns:
        platform_missing = df_clean[platform].isnull().sum()
        if platform_missing > 0:
            df_clean[platform] = df_clean[platform].fillna(False)
            print(f"   - {platform}: ç¼ºå¤± {platform_missing} è¡Œ -> å¡«å……ä¸º False")

print("\n" + "="*80)
print(f"âœ… ç¼ºå¤±å€¼å¤„ç†å®Œæˆ! å½“å‰æ•°æ®è¡Œæ•°: {len(df_clean)}")
```

## æ­¥éª¤5: æ•°æ®ç±»å‹è½¬æ¢å’Œè§„èŒƒåŒ–

```python
print("ğŸ”„ æ•°æ®ç±»å‹è½¬æ¢å’Œè§„èŒƒåŒ–:")
print("="*80)

# 1. è½¬æ¢æ—¥æœŸæ ¼å¼ (Release date)
print("\n1ï¸âƒ£ è½¬æ¢å‘å¸ƒæ—¥æœŸæ ¼å¼:")
if 'Release date' in df_clean.columns:
    print(f"   åŸå§‹æ•°æ®ç±»å‹: {df_clean['Release date'].dtype}")
    print(f"   æ ·ä¾‹æ•°æ®: {df_clean['Release date'].head(3).tolist()}")
    
    # è½¬æ¢ä¸ºdatetimeæ ¼å¼,é”™è¯¯çš„å€¼è®¾ä¸ºNaT
    df_clean['Release date'] = pd.to_datetime(df_clean['Release date'], errors='coerce')
    
    # ç»Ÿè®¡è½¬æ¢å¤±è´¥çš„æ•°é‡
    failed_conversion = df_clean['Release date'].isnull().sum()
    print(f"   è½¬æ¢åæ•°æ®ç±»å‹: {df_clean['Release date'].dtype}")
    print(f"   è½¬æ¢å¤±è´¥/ç¼ºå¤±: {failed_conversion} è¡Œ")
    
    # ä¸ºç¼ºå¤±çš„æ—¥æœŸå¡«å……ä¸€ä¸ªé»˜è®¤å€¼(å¦‚1970-01-01)æˆ–ä¿æŒNaN
    # è¿™é‡Œæˆ‘ä»¬ä¿æŒNaT,åç»­ä½¿ç”¨æ—¶å†å†³å®šå¦‚ä½•å¤„ç†
    print(f"   âœ… æ—¥æœŸè½¬æ¢å®Œæˆ")

# 2. è§„èŒƒåŒ–ä»·æ ¼å­—æ®µ (Price)
print("\n2ï¸âƒ£ è§„èŒƒåŒ–ä»·æ ¼å­—æ®µ:")
if 'Price' in df_clean.columns:
    print(f"   åŸå§‹æ•°æ®ç±»å‹: {df_clean['Price'].dtype}")
    
    # ç¡®ä¿ä»·æ ¼æ˜¯æ•°å€¼å‹
    df_clean['Price'] = pd.to_numeric(df_clean['Price'], errors='coerce')
    
    # å¤„ç†å¼‚å¸¸ä»·æ ¼(è´Ÿå€¼è®¾ä¸º0)
    negative_prices = (df_clean['Price'] < 0).sum()
    if negative_prices > 0:
        print(f"   å‘ç° {negative_prices} ä¸ªè´Ÿä»·æ ¼,å·²è®¾ä¸º0")
        df_clean.loc[df_clean['Price'] < 0, 'Price'] = 0
    
    # å¡«å……è½¬æ¢å¤±è´¥çš„å€¼ä¸º0
    df_clean['Price'] = df_clean['Price'].fillna(0)
    
    print(f"   ä»·æ ¼èŒƒå›´: ${df_clean['Price'].min():.2f} - ${df_clean['Price'].max():.2f}")
    print(f"   å¹³å‡ä»·æ ¼: ${df_clean['Price'].mean():.2f}")
    print(f"   âœ… ä»·æ ¼è§„èŒƒåŒ–å®Œæˆ")

# 3. è½¬æ¢è¯„åˆ†å­—æ®µä¸ºæ•´æ•°ç±»1å‹
print("\n3ï¸âƒ£ è½¬æ¢è¯„åˆ†å­—æ®µ:")
for col in ['Positive', 'Negative', 'Recommendations']:
    if col in df_clean.columns:
        df_clean[col] = pd.to_numeric(df_clean[col], errors='coerce').fillna(0).astype(int)
        print(f"   - {col}: è½¬æ¢ä¸ºæ•´æ•°ç±»å‹ -> èŒƒå›´ [{df_clean[col].min()}, {df_clean[col].max()}]")

# 4. å¤„ç†æ··åˆç±»å‹çš„åˆ— - Estimated owners
print("\n4ï¸âƒ£ å¤„ç†Estimated ownerså­—æ®µ:")
if 'Estimated owners' in df_clean.columns:
    print(f"   åŸå§‹æ•°æ®ç±»å‹: {df_clean['Estimated owners'].dtype}")
    print(f"   æ ·ä¾‹æ•°æ®: {df_clean['Estimated owners'].head(3).tolist()}")
    
    # å°†å­—ç¬¦ä¸²è½¬æ¢ä¸ºå­—ç¬¦ä¸²ç±»å‹(ç»Ÿä¸€æ ¼å¼)
    df_clean['Estimated owners'] = df_clean['Estimated owners'].astype(str)
    print(f"   âœ… è½¬æ¢ä¸ºå­—ç¬¦ä¸²ç±»å‹")

# 5. è½¬æ¢å¹³å°å­—æ®µä¸ºå¸ƒå°”å‹
print("\n5ï¸âƒ£ è½¬æ¢å¹³å°æ”¯æŒå­—æ®µ:")
for platform in ['Windows', 'Mac', 'Linux']:
    if platform in df_clean.columns:
        # è½¬æ¢ä¸ºå¸ƒå°”ç±»å‹
        df_clean[platform] = df_clean[platform].astype(bool)
        support_count = df_clean[platform].sum()
        print(f"   - {platform}: {support_count} ä¸ªæ¸¸æˆæ”¯æŒ ({support_count/len(df_clean)*100:.1f}%)")

# 6. è½¬æ¢Metacritic scoreä¸ºæ•°å€¼å‹
print("\n6ï¸âƒ£ è½¬æ¢Metacritic score:")
if 'Metacritic score' in df_clean.columns:
    df_clean['Metacritic score'] = pd.to_numeric(df_clean['Metacritic score'], errors='coerce').fillna(0).astype(int)
    non_zero = (df_clean['Metacritic score'] > 0).sum()
    print(f"   æœ‰Metacriticè¯„åˆ†çš„æ¸¸æˆ: {non_zero} ä¸ª ({non_zero/len(df_clean)*100:.1f}%)")
    if non_zero > 0:
        print(f"   è¯„åˆ†èŒƒå›´: {df_clean[df_clean['Metacritic score']>0]['Metacritic score'].min()} - {df_clean['Metacritic score'].max()}")
    print(f"   âœ… è½¬æ¢å®Œæˆ")

# 7. è½¬æ¢æ¸¸æˆæ—¶é•¿å­—æ®µä¸ºæ•°å€¼å‹
print("\n7ï¸âƒ£ è½¬æ¢æ¸¸æˆæ—¶é•¿å­—æ®µ:")
for playtime_col in ['Average playtime forever', 'Average playtime two weeks',
                      'Median playtime forever', 'Median playtime two weeks']:
    if playtime_col in df_clean.columns:
        df_clean[playtime_col] = pd.to_numeric(df_clean[playtime_col], errors='coerce').fillna(0)
        print(f"   - {playtime_col}: å¹³å‡ {df_clean[playtime_col].mean():.2f} åˆ†é’Ÿ")

# 8. è½¬æ¢å…¶ä»–è¾…åŠ©å­—æ®µ
print("\n8ï¸âƒ£ è½¬æ¢å…¶ä»–è¾…åŠ©å­—æ®µ:")
# Peak CCU
if 'Peak CCU' in df_clean.columns:
    df_clean['Peak CCU'] = pd.to_numeric(df_clean['Peak CCU'], errors='coerce').fillna(0).astype(int)
    print(f"   - Peak CCU: æœ€å¤§å€¼ {df_clean['Peak CCU'].max():,}")

# Required age
if 'Required age' in df_clean.columns:
    df_clean['Required age'] = pd.to_numeric(df_clean['Required age'], errors='coerce').fillna(0).astype(int)
    print(f"   - Required age: èŒƒå›´ {df_clean['Required age'].min()} - {df_clean['Required age'].max()}")

# Supported languages - ä¿æŒå­—ç¬¦ä¸²ç±»å‹
if 'Supported languages' in df_clean.columns:
    df_clean['Supported languages'] = df_clean['Supported languages'].astype(str)
    print(f"   - Supported languages: è½¬æ¢ä¸ºå­—ç¬¦ä¸²ç±»å‹")

print("\n" + "="*80)
print(f"âœ… æ•°æ®ç±»å‹è½¬æ¢å®Œæˆ!")
```
